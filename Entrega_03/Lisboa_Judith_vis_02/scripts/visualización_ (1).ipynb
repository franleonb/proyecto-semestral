```python
import pandas as pd
import altair as alt

file_path = 'Basedata_proyectro_ipc_grafica.csv'
df = pd.read_csv(file_path)

mapa_grupos = {
    'LECHE LÍQUIDA': 'Leche',
    'LECHE VEGANA': 'Leche',
    'QUESOS': 'Queso',
    'QUESO VEGANOS': 'Queso',
    'YOGURES Y PRODUCTOS SIMILARES': 'Yogurt',
    'YOGURT VEGANO': 'Yogurt',
    'MANTEQUILLA': 'Mantequilla',
    'MANTEQUILLA DE ACEITE DE COCO': 'Mantequilla',
    'MANJAR Y DULCES UNTABLES': 'Manjar',
    'MANJAR VEGANO': 'Manjar',
    'HELADOS': 'Helado',
    'HELADO VEGANO': 'Helado',
    'CARNE DE VACUNO': 'Proteína (Carne vs Tofu)',
    'TOFU': 'Proteína (Carne vs Tofu)',
    'PAN': 'Pan',
    'PAN VEGANO': 'Pan',
    'GALLETAS': 'Galletas',
    'GALLETAS VEGANAS': 'Galletas',
    'PASTAS': 'Pastas',
    'PASTAS SIN HUEVO': 'Pastas',
    'HAMBURGESA': 'Hamburguesa',
    'HAMBURGUESA VEGANA': 'Hamburguesa',
}

# Filtrar solo los productos que están en el mapa
productos_a_filtrar = list(mapa_grupos.keys())
df_filtrado = df[df['PRODUCTOS'].isin(productos_a_filtrar)]

source = df_filtrado[['PRODUCTOS', 'Tipo de dieta', 'oct-25']].copy()

# Crear la columna 'Grupo' (esta será nuestra 'row' o 'site')
source['Grupo'] = source['PRODUCTOS'].map(mapa_grupos)

# Renombrar 'oct-25' a 'Precio' and 'Tipo de dieta' to 'Dieta'
source = source.rename(columns={'oct-25': 'Precio', 'Tipo de dieta': 'Dieta'})

chart = alt.Chart(source).mark_bar().encode(

# Eje X (el largo de la barra):
# 'Precio:Q' (equivale a 'x=sum(yield):Q')
x=alt.X('Precio:Q', title='Precio (Oct 2025)', axis=alt.Axis(format='$,.0f')),

# Eje Y (las barras):
# 'Dieta:O' (equivale a 'y=year:O')
y=alt.Y('Dieta:O', title=None, sort=None, axis=None),
# Ocultamos la etiqueta del eje Y

# Color (el color de las barras):
# 'Dieta:N' (equivale a 'color=year:N')
color=alt.Color('Dieta:N', title="Tipo de Dieta",
scale=alt.Scale(domain=['Tradicional', 'Vegano'],
range=['#7FB3D5', '#E74C3C']), legend=alt.Legend(
title="Tipo de Dieta",
labelExpr="datum.label == 'Tradicional' ? 'Tradicional / Lobo' : 'Vegano / Caperucita'"

)),

# Fila (la agrupación principal):
# 'Grupo:N' (equivale a 'row=site:N')
row=alt.Row('Grupo:N', header=alt.Header(
        title="Comparativa de Productos (Tradicional vs. Vegano)",
        titleOrient="top",
        labelOrient="left",
        labelAngle=0,
        labelAlign='left',
        labelPadding=5
    )),

# Tooltip: Información al pasar el mouse
tooltip=['Grupo', 'PRODUCTOS', 'Dieta', 'Precio']

).properties(
    title=alt.Title("¿Dónde se esconde el costo?: Caperucita vs. Lobo",
    anchor="middle")

  # Centered the title
)

chart
```

```python
chart.save('visualización.html')
```
