```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd

df = pd.read_csv(next(iter(uploaded)))
df.head()
```

```python
descripciones = {
    'MANTENIDOS': 'Productos que no cambian entre canastas',
    'REEMPLAZADOS': 'Productos que se sustituyen por opción vegana',
    'ELIMINADOS': 'Productos que desaparecen en la canasta vegana',
    'AGREGADOS': 'Productos nuevos que aparecen en la canasta vegana',
    'EXTENDIDOS': 'Un producto se divide en varios',
    'DISMINUIDOS': 'Varios productos se consolidan en uno solo'
}

df['Descripcion'] = df['CATEGORÍA'].map(descripciones)
df.head()
```

```python
import altair as alt

chart = alt.Chart(df).mark_bar().encode(
    x=alt.X('CATEGORÍA:N', title='Tipo de cambio de producto'),
    y=alt.Y('CANTIDAD:Q', title='Número de productos'),
    color=alt.Color('CATEGORÍA:N', legend=alt.Legend(title="Categoría de cambio")),
    tooltip=['CATEGORÍA', 'CANTIDAD', 'Descripcion']
).properties(
    title='Cambios en los productos: Canasta tradicional vs vegana',
    width=600,
    height=400
).interactive()

chart
```

```python
chart.save('visualizacion.html')
print("Archivo guardado: visualizacion.html")
```
