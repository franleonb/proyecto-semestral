```python
from google.colab import files
import pandas as pd
import plotly.express as px

uploaded = files.upload()
df = pd.read_csv(next(iter(uploaded)))
```

```python
df = df[['SUPERMERCADO', 'DISPONIBILIDAD']].dropna()

df[['supermercado', 'estado']] = df['SUPERMERCADO'].str.split(' - ', expand=True)

df['DISPONIBILIDAD'] = pd.to_numeric(df['DISPONIBILIDAD'], errors='coerce')

df['grupo'] = df['supermercado'] + ' - ' + df['estado']

color_map = {
    'TOTTUS - DISPONIBLES': '#d90429',
    'TOTTUS - NO DISPONIBLES': '#f9b5b5',
    'LIDER - DISPONIBLES': '#0057d9',
    'LIDER - NO DISPONIBLES': '#a9c9ff',
    'JUMBO - DISPONIBLES': '#008000',
    'JUMBO - NO DISPONIBLES': '#98e6a6'
}

fig = px.bar(
    df,
    x='supermercado',
    y='DISPONIBILIDAD',
    color='grupo',
    barmode='group',
    title='Disponibilidad de productos veganos por supermercado',
    color_discrete_map=color_map
)

fig.update_layout(
    xaxis_title="Supermercado",
    yaxis_title="Cantidad de productos",
    legend_title="Categoría"
)

fig.show()
```

```python
fig.write_html('vis-productos-final.html')
```
